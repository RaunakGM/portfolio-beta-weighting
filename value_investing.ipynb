```python
import numpy as np
import pandas as pd
import math
import xlsxwriter
from scipy import stats
```

```python
df=pd.read_csv("value_investing.csv")
df
```

```python
df['P/E percentile']=np.nan
df['P/BV percentile']=np.nan
df['P/FCF percentile']=np.nan
df['Value']=np.nan
```

```python
df.columns
```

```python
metrics={'P/E RATIO': 'P/E percentile',
         'P/BV':'P/BV percentile',
         'P/FCF':'P/FCF percentile'
         
         }
for metric in metrics.keys():
    for row in df.index:
        df.loc[row,metrics[metric]]=stats.percentileofscore(df[metric],df.loc[row,metric])
df
```

```python
from statistics import mean
for row in df.index:
    pe_ratio=df.loc[row,'P/E percentile']
    pbv_ratio=df.loc[row,'P/BV percentile']
    pfcf_ratio=df.loc[row,'P/FCF percentile']
 

    df.loc[row,'Value'] = (pe_ratio + pbv_ratio + pfcf_ratio)/3
df
```

```python
df.sort_values('Value',inplace=True)
df.reset_index(drop=True,inplace=True)
df
```
